```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense,SimpleRNN
import numpy as np
import nltk
from nltk.corpus import gutenberg
import re
```

```python
nltk.download('punkt')
nltk.download('gutenberg')
```

```python
# load the preprocess data
text = gutenberg.raw('shakespeare-hamlet.txt').lower()
text = re.sub(r'[^a-z\s]', '', text)
tokens = nltk.word_tokenize(text)

#parameters
vocab_size = 10000
embedding_dim = 50
max_length = 5 # Context size

#Tokenize
tokenizer = Tokenizer(num_words=vocab_size, oov_token='<OOV>')
tokenizer.fit_on_texts(tokens)
sequences = tokenizer.texts_to_sequences([tokens])[0]
print(sequences)
```

```python
#create input-output pairs
X = []
y = []
for i in range(len(sequences) - max_length):
    X.append(sequences[i:i+max_length])
    y.append(sequences[i+max_length])
X = np.array(X)
y = np.array(y)
```

```python
# Build the model
model =  Sequential([
    Embedding(vocab_size, embedding_dim, input_length=max_length),
    SimpleRNN(100, return_sequences=False),
    Dense(vocab_size, activation='softmax')
])
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
# Train the model
model.fit(X, y, epochs=50, batch_size=64 , validation_split=0.8)
```

```python
# Function to generate next word
def generate_next_word(model, tockenizer, seed_text, next_words):
  seed_tockens = tokenizer.texts_to_sequences([seed_text.lower])[0]
  seed_tockens = seed_tockens[-max_length:]
  pad_seed = pad_sequences([seed_tockens], maxlen=max_length, padding='pre')
  pred = model.predict(pad_seed,verbose=0)
  next_id = np.argmax(pred,axixs=1)[0]
  return tokenizer.index_word.get[next_id,'']
```

```python
# Example usage
seed = "to be or not "
next_words = generate_next_word(model, tokenizer, seed, max_length)
print(f"seed: '{seed}' -> Next word: '{next_words}'")
```
